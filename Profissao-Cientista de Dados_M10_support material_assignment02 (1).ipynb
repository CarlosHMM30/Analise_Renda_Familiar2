````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

arquivo = pd.read_csv('previsao_de_renda.csv')
````

````python
df = pd.DataFrame(arquivo)
df
````

````python
df['posse_de_veiculo'] = df['posse_de_veiculo'].map({True:1,False:0})	
df['posse_de_imovel'] = df['posse_de_imovel'].map({True:1,False:0})	
df
````

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda', data=df)
````

````python
ax = sns.pointplot(x='posse_de_imovel', y='renda', data=df)
````
